```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("REALQUAMI_Dataset_Merged.csv")

# 1. Count frequencies for Condition_primary and Condition_secondary (descending)
cond1_counts = df["Condition_primary"].value_counts().sort_values(ascending=False)
cond2_counts = df["Condition_secondary"].value_counts().sort_values(ascending=False)

# 2. Convert Condition_secondary counts to negative so the bars face left
cond2_counts_neg = -cond2_counts

# 3. Prepare figure and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 8), sharey=True)

# --- LEFT CHART: Condition_primary (descending) ---
axes[0].barh(cond1_counts.index, cond1_counts.values, color="skyblue")
axes[0].set_title(f"Condition_primary (n={cond1_counts.sum()})", fontsize=12)
axes[0].set_xlabel("Count")
axes[0].invert_yaxis()  # So largest category is at top

# Annotate each bar with the count
for patch in axes[0].patches:
    width = patch.get_width()
    y_pos = patch.get_y() + patch.get_height() / 2
    # place the text just to the right of the bar
    axes[0].annotate(f"{int(width)}", (width, y_pos),
                     xytext=(5, 0), textcoords="offset points",
                     va='center', ha='left', fontsize=9)

# --- RIGHT CHART: Condition_secondary (mirrored) ---
axes[1].barh(cond2_counts_neg.index, cond2_counts_neg.values, color="salmon")
axes[1].set_title(f"Condition_secondary (n={cond2_counts.sum()})", fontsize=12)
axes[1].set_xlabel("Count")

# For the mirrored chart, we want largest category at top
axes[1].invert_yaxis()  # ensures the top is the largest category
axes[1].yaxis.tick_right()  # move y-axis labels to the right side
axes[1].yaxis.set_label_position("right")

# Annotate each bar with the absolute count
for patch in axes[1].patches:
    width = patch.get_width()  # negative value
    y_pos = patch.get_y() + patch.get_height() / 2
    # place text to the left of the bar (which extends leftward)
    axes[1].annotate(f"{int(abs(width))}", (width, y_pos),
                     xytext=(-5, 0), textcoords="offset points",
                     va='center', ha='right', fontsize=9)

# Remove the default y-label from the second chart
axes[1].set_ylabel("")

# Adjust layout so there's space in the middle for category labels
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# -------------------------
# 1. LOAD DATA
# -------------------------
df = pd.read_csv("REALQUAMI_Dataset_Merged.csv")
total_cases = len(df)

# Count frequencies
cond1_counts = df["Condition_primary"].value_counts().sort_values(ascending=False)
cond2_counts = df["Condition_secondary"].value_counts().sort_values(ascending=False)

# Convert second chart's counts to negative for mirroring
cond2_neg = -cond2_counts

# -------------------------
# 2. PREP FIGURE & AXES
# -------------------------
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 8), sharey=True)

# We'll place bars/annotations on top of the grid lines (zorder).
# The grid lines will appear behind them.

# -------------------------
# 3. LEFT CHART (Condition_primary)
# -------------------------
axes[0].barh(cond1_counts.index, cond1_counts.values, color="skyblue", zorder=2)
axes[0].set_title(f"Condition_primary (n={cond1_counts.sum()}, total={total_cases})", fontsize=11)
axes[0].set_xlabel("Count")
axes[0].invert_yaxis()  # largest category at top

# Annotate each bar with its count
for patch in axes[0].patches:
    width = patch.get_width()
    y_center = patch.get_y() + patch.get_height() / 2
    axes[0].annotate(
        f"{int(width)}",
        (width, y_center),
        xytext=(5, 0),      # offset 5 px to the right
        textcoords="offset points",
        va='center', ha='left', fontsize=9,
        zorder=3            # ensure annotation is on top
    )

# -------------------------
# 4. RIGHT CHART (Condition_secondary, mirrored)
# -------------------------
axes[1].barh(cond2_neg.index, cond2_neg.values, color="salmon", zorder=2)
axes[1].set_title(f"Condition_secondary (n={cond2_counts.sum()}, total={total_cases})", fontsize=11)
axes[1].set_xlabel("Count")
axes[1].invert_yaxis()         # largest category at top
axes[1].yaxis.tick_right()     # move y labels to right side
axes[1].yaxis.set_label_position("right")

# Annotate each bar (negative) with absolute count
for patch in axes[1].patches:
    width = patch.get_width()  # negative
    y_center = patch.get_y() + patch.get_height() / 2
    axes[1].annotate(
        f"{int(abs(width))}",
        (width, y_center),
        xytext=(-5, 0),   # offset 5 px to the left
        textcoords="offset points",
        va='center', ha='right', fontsize=9,
        zorder=3
    )

# -------------------------
# 5. SHOW GRID LINES FOR EVERY 3RD SUBVARIABLE
# -------------------------
# We'll define major ticks every 3 subvariables, minor ticks for each subvariable,
# then draw dotted grid lines only for major ticks.

# Condition_primary indexes
cond1_labels = cond1_counts.index.tolist()
num_cond1 = len(cond1_labels)

# Condition_secondary indexes
cond2_labels = cond2_counts.index.tolist()
num_cond2 = len(cond2_labels)

# For sharey to work well, we want the same number of total categories on y-axis.
# We'll unify them into the max of the two sets for consistent ticks.
max_categories = max(num_cond1, num_cond2)

# Set major locator for every 3rd index
axes[0].yaxis.set_major_locator(ticker.MultipleLocator(3))
axes[1].yaxis.set_major_locator(ticker.MultipleLocator(3))

# Minor locator for every single index
axes[0].yaxis.set_minor_locator(ticker.MultipleLocator(1))
axes[1].yaxis.set_minor_locator(ticker.MultipleLocator(1))

# We'll keep the default labels from the bar plot, but we want to ensure the grid lines
# appear only on major ticks:
axes[0].grid(axis='y', which='major', linestyle='--', alpha=0.6, color='grey', zorder=1)
axes[1].grid(axis='y', which='major', linestyle='--', alpha=0.6, color='grey', zorder=1)

# We unify the x-limits for neat alignment
max_left = cond1_counts.max()
max_right = cond2_counts.max()
axes[0].set_xlim(0, max_left * 1.1)
axes[1].set_xlim(-max_right * 1.1, 0)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("REALQUAMI_Dataset_Merged.csv")

# Count frequencies for the two condition variables (sorted descending)
cond1_counts = df["Condition_primary"].value_counts().sort_values(ascending=False)
cond2_counts = df["Condition_secondary"].value_counts().sort_values(ascending=False)

# For the right chart, we convert counts to negative so the bars extend leftwards (mirrored)
cond2_neg = -cond2_counts

# Prepare figure with two subplots sharing the same y-axis
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 8), sharey=True)

# --------------------
# Left Chart: Condition_primary
# --------------------
# We'll plot the bars manually using the index positions so we can ensure all labels appear
y_positions_1 = range(len(cond1_counts))
axes[0].barh(y_positions_1, cond1_counts.values, color="skyblue", zorder=2)
axes[0].set_title("Condition_primary", fontsize=12)
axes[0].set_xlabel("Count")
axes[0].set_yticks(y_positions_1)
axes[0].set_yticklabels(cond1_counts.index)
axes[0].invert_yaxis()  # So the largest category appears at the top

# Annotate each bar on the left chart
for i, value in enumerate(cond1_counts.values):
    axes[0].annotate(f"{int(value)}",
                     (value, i),
                     xytext=(5, 0),  # offset 5 pixels to the right
                     textcoords="offset points",
                     va="center", ha="left", fontsize=9, zorder=3)

# --------------------
# Right Chart: Condition_secondary
# --------------------
y_positions_2 = range(len(cond2_counts))
axes[1].barh(y_positions_2, cond2_neg.values, color="salmon", zorder=2)
axes[1].set_title("Condition_secondary", fontsize=12)
axes[1].set_xlabel("Count")
axes[1].set_yticks(y_positions_2)
axes[1].set_yticklabels(cond2_counts.index)
axes[1].invert_yaxis()  # Largest category at the top
axes[1].yaxis.tick_right()  # Move y-axis labels to the right side
axes[1].yaxis.set_label_position("right")

# Annotate each bar on the right chart (using absolute count)
for i, value in enumerate(cond2_counts.values):
    axes[1].annotate(f"{int(value)}",
                     (-value, i),
                     xytext=(-5, 0),  # offset 5 pixels to the left
                     textcoords="offset points",
                     va="center", ha="right", fontsize=9, zorder=3)

# --------------------
# Add Horizontal Dotted Grid Lines Every 3rd Category
# --------------------
# Here, we manually draw horizontal lines at positions corresponding to every third label.
for i in range(0, max(len(cond1_counts), len(cond2_counts)), 3):
    # The y-coordinate for a horizontal line is set at (i - 0.5) to fall between bars.
    axes[0].axhline(i - 0.5, color="grey", linestyle="--", alpha=0.6, zorder=1)
    axes[1].axhline(i - 0.5, color="grey", linestyle="--", alpha=0.6, zorder=1)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load data
df = pd.read_csv("REALQUAMI_Dataset_Merged.csv")

# 2. Get frequencies for Condition_primary (descending) and Condition_secondary (descending)
cond1_counts = df["Condition_primary"].value_counts().sort_values(ascending=False)
cond2_counts = df["Condition_secondary"].value_counts().sort_values(ascending=False)

# 3. Create a combined index that ensures:
#    - All categories from Condition_primary in descending order
#    - Then any extra categories from Condition_secondary appended below
cond1_index = cond1_counts.index
cond2_index = cond2_counts.index

# Categories present in Condition_secondary but not in Condition_primary
extra_cats = cond2_index.difference(cond1_index)

# Sort those extra categories by their frequency in cond2_counts (descending)
extra_cats_sorted = cond2_counts.loc[extra_cats].sort_values(ascending=False).index

# Final index: cond1_index on top, then extra_cats_sorted
final_index = cond1_index.append(extra_cats_sorted)

# 4. Reindex both counts so they share exactly the same index
cond1_aligned = cond1_counts.reindex(final_index, fill_value=0)
cond2_aligned = cond2_counts.reindex(final_index, fill_value=0)

# 5. Convert Condition_secondary to negative for mirroring
cond2_neg = -cond2_aligned

# 6. Create subplots with sharey=True so subvariable names align
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 8), sharey=True)

# -----------------------------
# LEFT CHART: Condition_primary
# -----------------------------
axes[0].barh(cond1_aligned.index, cond1_aligned.values, color="skyblue", zorder=2)
axes[0].set_title("Condition_primary (n={cond1_counts.sum()}", fontsize=12)
axes[0].set_xlabel("Count")

# Largest category on top
axes[0].invert_yaxis()

# Annotate each bar with count
for patch in axes[0].patches:
    width = patch.get_width()
    y_center = patch.get_y() + patch.get_height()/2
    axes[0].annotate(
        f"{int(width)}",
        (width, y_center),
        xytext=(5, 0),
        textcoords="offset points",
        va="center", ha="left", fontsize=9
    )

# -----------------------------
# RIGHT CHART: Condition_secondary (mirrored)
# -----------------------------
axes[1].barh(cond2_neg.index, cond2_neg.values, color="salmon", zorder=2)
axes[1].set_title("Condition_secondary (n={cond2_counts.sum()}", fontsize=12)
axes[1].set_xlabel("Count")

# Largest category on top
axes[1].invert_yaxis()

# Move y-axis labels to the right side
axes[1].yaxis.tick_right()
axes[1].yaxis.set_label_position("right")

# Annotate each bar (negative) with absolute count
for patch in axes[1].patches:
    width = patch.get_width()  # negative
    y_center = patch.get_y() + patch.get_height()/2
    axes[1].annotate(
        f"{int(abs(width))}",
        (width, y_center),
        xytext=(-5, 0),
        textcoords="offset points",
        va="center", ha="right", fontsize=9
    )

# OPTIONAL: Add dotted horizontal lines every 3rd category
for i in range(0, len(final_index), 3):
    # i - 0.5 to draw a line between categories
    axes[0].axhline(i - 0.5, color="grey", linestyle="--", alpha=0.6, zorder=1)
    axes[1].axhline(i - 0.5, color="grey", linestyle="--", alpha=0.6, zorder=1)

# Align x-limits for a consistent mirrored look
max_left = cond1_aligned.max()
max_right = cond2_aligned.max()
axes[0].set_xlim(0, max_left * 1.1)
axes[1].set_xlim(-max_right * 1.1, 0)

plt.tight_layout()
plt.show()
```
