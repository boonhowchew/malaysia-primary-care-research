```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (adjust filename/path if needed)
df = pd.read_csv("REALQUAMI_Dataset_Merged.csv")

# Show first few rows to confirm loading
df.head()
```

```python
# 1. Count how many times each condition appears, sorting by descending
cond1_counts = df["Condition_primary"].value_counts().sort_values(ascending=False)
cond2_counts = df["Condition_secondary"].value_counts().sort_values(ascending=False)

# 2. Create a figure with two subplots side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --- Left chart: Condition_primary ---
sns.barplot(
    x=cond1_counts.values,
    y=cond1_counts.index,
    order=cond1_counts.index,
    ax=axes[0],
    palette="Blues_d"
)
axes[0].set_title("Condition_primary (Descending)")
axes[0].set_xlabel("Count")
axes[0].set_ylabel("Condition_primary")

# --- Right chart: Condition_secondary ---
sns.barplot(
    x=cond2_counts.values,
    y=cond2_counts.index,
    order=cond2_counts.index,
    ax=axes[1],
    palette="Reds_d"
)
axes[1].set_title("Condition_secondary (Descending)")
axes[1].set_xlabel("Count")
axes[1].set_ylabel("Condition_secondary")

plt.tight_layout()
plt.show()
```
